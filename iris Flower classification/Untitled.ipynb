```python
#scatter plot
colors = ['red','orange','blue']
species = ['Iris-setosa','Iris-versicolor','Iris-virginica']
```

```python
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['SepalLengthCm'],x['SepalWidthCm'],c=colors[i],label=species[i])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
##way too much scattered
```

```python
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['PetalLengthCm'],x['PetalWidthCm'],c=colors[i],label=species[i])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
#better
```

```python
##label encoder conversts string to numeric possible form bruh
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
##converting species column to computer readable form
df['Species'] = le.fit_transform(df['Species']) #it basically converts names into numerics labels 
df.head()
```

```python
##finnally model training time
from sklearn.model_selection import train_test_split
#train,test data spltting
X = df.drop(columns = ['Species']) #droping species as its y
Y = df['Species']
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.30)
```

```python
#will  try to use logistic regression
from sklearn.linear_model import LogisticRegression ##its a classification
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test) ## its not a good one as 100 accuracy
```

```python
#another model
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
#decison tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
